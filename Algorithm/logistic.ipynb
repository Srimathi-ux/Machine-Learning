```python
df = pd.read_csv("/content/HR_comma_sep.csv")
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='left', data=df)
plt.title('Employee Retention by Department')
plt.xlabel('Department')
plt.ylabel('Retention Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='salary', y='left', data=df, order=['low', 'medium', 'high'])
plt.title('Employee Retention by Salary Level')
plt.xlabel('Salary Level')
plt.ylabel('Retention Rate')
plt.tight_layout()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['salary']],df.left,train_size=0.8)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
X_train_encoded = pd.get_dummies(X_train, columns=['salary'], drop_first=True)
X_test_encoded = pd.get_dummies(X_test, columns=['salary'], drop_first=True)
model.fit(X_train_encoded, y_train)
```

```python
X_test
```

```python
y_predicted = model.predict(X_test_encoded)
```

```python
model.score(X_test_encoded, y_test)
```
